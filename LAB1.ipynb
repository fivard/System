```python
def GCD(a, b):
  a, b = sorted((a, b))
  while b != 0:
    a, b = b, a % b
  return a

def task_2():
  text = input('Enter 2 numbers: ')
  values = text.split(' ')
  assert len(values) == 2, f'Only 2 numbers allowed. Entered {len(values)}'
  try:
    values = [int(v) for v in values]
  except ValueError:
    raise ValueError('Only numbers allowed')
  
  assert all(v > 0 for v in values), "Only positive numvers allowed"
  gcd = GCD(*values)

  return f'GCD of {values} is {gcd}'

  
```
